```python
#fibo


def fibo_nly(n):
    if n ==0 :
        return 1
    if n== 1:
        return 1
    return fibo_nly(n-1)+fibo_nly(n-2)
    
fibo_nly(30)
```

```python
f=[0]*500
f[0] = 1
f[1] = 1
def fibo_dp(n):
    
    for i in range(2,n+1):
        
        f[i] = f[i-1]+f[i-2]
    return f[n]

fibo_dp(30)
```

```python
#fact_tabulation
f=[1]*300
f[0]=1
f[1]=1
def fact(n):
    
    for i in range(2,n+1):
        f[i]=i*f[i-1]
  
    return f[n]

fact(10)
    
```

```python
#code
import sys
matrix= [ [1, 2, 3], 
        [4, 8, 2], 
        [1, 5, 3] ] 

def findPath(point):
    r = point[0]
    c = point[1]
    if r<0 or c <0 :
        return sys.maxsize 
    if r ==0 and c == 0:
        return matrix[r][c]
    minnn = min(findPath((r-1,c-1)),findPath((r,c-1)),findPath((r-1,c)))
    
    #print("min",minnn)
    return matrix[r][c] + minnn 

point=[2,2]
res = findPath(point)
print(res)
```

```python
def minCost(cost, m,n): 
    if (n < 0 or m < 0): 
        return sys.maxsize 
    elif (m == 0 and n == 0): 
        return cost[m][n] 
    else: 
        minnn = min( minCost(cost, m-1, n-1), 
                                minCost(cost, m-1, n), 
                                minCost(cost, m, n-1) )
       
        return cost[m][n] + minnn  

point=[2,2]
res = minCost(matrix,point[0],point[1])
print(res)
```

```python
#catalan rec

def catalan_nly(n):
    
    mult=1
    for k in range(2,n+1):
        mult*=((n/k)+1)    
    return mult
    
    
catalan_nly(100)    
    

```

```python
def catalan_rec(n,k):
    
    if k == 1:
        return 1
    
    return (n/k + 1)*catalan_rec(n,k-1)
    
catalan_rec(100,100)
```

```python
def catalan_rec2(n):
    summ=0
    if n <= 1:
        return 1
    for i in range(n):
        
        summ += catalan_rec2(i)*catalan_rec2(n-i-1)
        
    return summ  
      
catalan_rec2(3)
        
```

```python
#LCS-three

def LCS_2(arr1,arr2,n,m):
    
    seq=""
    
    if arr1[n-1] == arr2[m-2]:
        return arr1[n-1] + LCS_2(arr1,arr2,n-1,m-1)
    
    return arr1[n-1] + max(LCS_2(arr1,arr2,n-1,m),LCS_2(arr1,arr2,n,m-1))

def LCS_3(arr1,arr2,arr3):
    
    return LCS_2(arr1,LCS_2(arr2,arr3))


```
