```python
# any sort

def any_sort(arr):
    
    for i in range(len(arr)):
        smallest = arr[i]
        change = i
        for j in range(i,len(arr)):            
            if arr[j] < smallest:                
                smallest = arr[j]
                change = j
        
        if smallest != arr[i]:
            arr[change], arr[i] = arr[i], arr[change]
    
    return arr
    
arr = [3,4,1,2,8,0]
print(arr)
print(any_sort(arr))


```

```python
#quick sort

def quick_sort(arr,low,high):
    
    if low >= high:
        return
    
    middle = partiton(arr,low,high)
    quick_sort(arr,low,middle-1)
    quick_sort(arr,middle+1,high)
    
def partiton(arr,low,high):
    
    head = low
    pivot = arr[high]
    
    for i in range(low,high-1):
        
        if arr[i] < pivot:
            arr[head],arr[i]=arr[i],arr[head]
            head += 1
            
    arr[high] ,arr[head] = arr[head], arr[high]
    return head

arr = [3,8,1,9,6]

quick_sort(arr,0,len(arr)-1)
print(arr)
    
```
